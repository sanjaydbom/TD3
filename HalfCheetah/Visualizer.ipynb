```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('HalfCheetahTestingData.csv')
plt.hist(df['score'], bins=1000, edgecolor='black')
plt.show()
```

```python
#remove outliers
mean_score = np.mean(df['score'])
std = np.std(df['score'])
scores = df[df['score'] > mean_score - 1.5 * std]
scores = scores[scores['score'] < mean_score + 1.5 * std]
print(len(scores['score']))
```
